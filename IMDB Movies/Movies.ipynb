# Movies per Genre

A bar plot is used to show the top 10 genres in the movies.

Some of the movies have multiple genres like "Schindler's List" has three genres: 
- Biography
- Drama
- History

So each of the genre is counted as one. So Drama has appear 9 times in the top 10 movies while Crime has appeared 5 times.

```python
all_genres = df['Genre'].str.split(',').explode().str.strip()
```

```python
genre_count = all_genres.value_counts()
```

```python
print(genre_count)
```

```python
print(genre_count.index)
```

```python
print(genre_count.values)
```

```python
colors = ['#4A90E2', '#50E3C2', "#98E051", '#F8E71C', '#F5A623', '#D0021B', '#9013FE', '#7ED321']

plt.figure(figsize=(10,6))
plt.bar(genre_count.index, genre_count.values, color=colors, edgecolor="black")
plt.ylim(0, 11)
plt.yticks([i for i in range(0, 11)])
plt.ylabel("Number of Movies")
plt.title("Movies per Genre")
plt.show()
```

# Genre Distribution

A pie chart is used to show the genre distribution.

So Drama being the most appeaed genre in the data represent 41% of the data.

```python
plt.figure(figsize=(8, 6))
plt.pie(
    genre_count.values,
    colors=colors,
    labels=genre_count.index,
    autopct="%1.0f%%",
    textprops={"color": "black", "fontsize": 10}
)

plt.title("Genre Distribution", fontsize=16, fontweight="bold")
plt.show()
```

# Release Year vs IMDB Rating

A scatter plot is used to compare the released year and rating.

The result shows that old movies have a higher IMDB Rating, maybe they are just better than most of today's films or the number of reviews are less.

```python
years = df["Released_Year"]
print(years)
```

```python
years= years.sort_values(ascending=True).values
print(years)
```

```python
ratings = df["IMDB_Rating"].values
print(ratings)
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(years, ratings, color="#FFE600", marker="*", s=200, edgecolors="black", linewidths=0.5)
plt.title("Year vs Rating", fontsize=16)
plt.xlabel("Released Year", fontsize=12)
plt.ylabel("IMDB Ratings", fontsize=12)
plt.yticks(np.arange(8.5, 9.6, 0.1))
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

# Movies Per Decade

A bar plot is used to show movies per decade. The released years are first coverted into decades like 1994 --> 1990.

Then counted, how many movies in each decade.

The result shows that the one of the top movies were released in the decade 1990s.

```python
# Make a copy of dataframe
df2 = df.copy()

# Convert year --> decade (e.g., 1994 --> 1990s)
df2['Released_Year'] = df2['Released_Year'].astype(int)

df2['Decade'] = (df2['Released_Year'] // 10) * 10
df2['Decade'] = df2['Decade'].astype(str) + "s"

# Count movies per decade
decade_counts = df2['Decade'].value_counts()
```

```python
decade_counts
```

```python
decade_counts = decade_counts.sort_index()
```

```python
decade_counts
```

```python
plt.figure(figsize=(8,5))
colors = ['#4A90E2', '#F5A623', '#7ED321', '#9013FE', '#D0021B']

plt.bar(decade_counts.index, decade_counts.values,
        color=colors[:len(decade_counts)], edgecolor="black")

plt.xlabel("Decade", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.yticks([0, 1, 2, 3, 4])
plt.title("Movies per Decade", fontsize=16, fontweight="bold")
plt.show()
```

# Average IMDB Rating per Director

A simple line plot is used to show average IMDB Rating per director. Average is used because there are two repeated directoirs in the data.

```python
director_avg = df.groupby("Director")["IMDB_Rating"].mean().sort_values()
```

```python
director_avg
```

```python
plt.figure(figsize=(9,5))

plt.plot(director_avg.index,
         director_avg.values,
         marker="s",
         linestyle="-.",
         color="crimson")

plt.xlabel("Director")
plt.ylabel("Average IMDB Rating")
plt.title("Average IMDB Rating per Director")

plt.xticks(rotation=45, ha="right")
plt.grid(True, linestyle=":")
```
